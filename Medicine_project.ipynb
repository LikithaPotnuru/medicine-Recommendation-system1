# Drop duplicates

```python
# Check how many duplicates exist
print("Total duplicate rows:", df.duplicated().sum())

# Drop duplicate rows
df = df.drop_duplicates()

# Confirm duplicates are removed
print("Duplicates removed. Current rows:", df.shape[0])
```

# Missing values

```python
# Check missing values per column
missing_counts = df.isnull().sum()
missing_percentage = (missing_counts / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Percentage': missing_percentage
}).sort_values(by='Missing Count', ascending=False)

missing_df
```

# Drop missing values

```python
# Drop mostly empty sideEffect columns
side_effect_cols_to_drop = [f'sideEffect{i}' for i in range(6, 42)]
df = df.drop(columns=[col for col in side_effect_cols_to_drop if col in df.columns])

print("Remaining columns after dropping mostly empty sideEffect columns:")
print(df.columns.tolist())
```

# combining columns

```python
# Combine 'use' columns
use_cols = [f'use{i}' for i in range(5)]
df['uses'] = df[use_cols].apply(lambda x: '; '.join(x.dropna().astype(str)), axis=1)

# Combine 'sideEffect' columns
side_effect_cols = [f'sideEffect{i}' for i in range(6)]
df['side_effects'] = df[side_effect_cols].apply(lambda x: '; '.join(x.dropna().astype(str)), axis=1)

# Combine 'substitute' columns
substitute_cols = [f'substitute{i}' for i in range(5)]
df['substitutes'] = df[substitute_cols].apply(lambda x: '; '.join(x.dropna().astype(str)), axis=1)

# Quick check
df[['name', 'uses', 'side_effects', 'substitutes']].head(3)
```

# Drop old individual columns and keep final useful columns

```python
# Drop old columns
old_cols = use_cols + side_effect_cols + substitute_cols
df = df.drop(columns=old_cols)

# Keep only final useful columns
final_cols = ['id', 'name', 'Chemical Class', 'Therapeutic Class', 'Action Class', 'Habit Forming',
              'uses', 'side_effects', 'substitutes']

df = df[final_cols]

# Quick check
print("Final columns:", df.columns.tolist())
df.head(3)
```

```python
# Fill missing category values with 'Unknown'
df['Chemical Class'] = df['Chemical Class'].fillna('Unknown')
df['Action Class'] = df['Action Class'].fillna('Unknown')

# Quick check
df[['name', 'Chemical Class', 'Action Class']].head(5)
```

# Build a Simple Search Function


```python
# Step 2 — Simple search by medicine name

def search_medicine(query, df):
    # Convert both to lowercase for case-insensitive search
    result = df[df['name'].str.contains(query, case=False, na=False)]
    return result[['name', 'uses', 'side_effects', 'substitutes']].head(5)

# 🔍 Example test
query = "augmentin"
print(f"\nSearch results for: {query}")
display(search_medicine(query, df))
```

# Smarter Search (by name + uses/symptoms)

```python
# Step 3 — Smarter search (search in name + uses)

def smart_search(query, df):
    query = query.lower()

    results = df[
        df['name'].str.contains(query, case=False, na=False) |
        df['uses'].str.contains(query, case=False, na=False)
    ]

    return results[['name', 'uses', 'side_effects', 'substitutes']].head(10)

# 🔍 Example test searches
for test_query in ["augmentin", "cough", "allergy"]:
    print(f"\nSearch results for: {test_query}")
    display(smart_search(test_query, df))
```

# Add Simple Recommendation Logic

```python
# Step 4 — Recommendation: show substitutes as recommendations

def recommend_medicines(query, df):
    results = smart_search(query, df)

    recommendations = []
    for _, row in results.iterrows():
        subs = str(row['substitutes']).split(";")
        subs = [s.strip() for s in subs if pd.notna(s)]
        recommendations.extend(subs)

    recommendations = list(set(recommendations))[:10]  # unique & limit
    return results, recommendations


# 🔍 Example test
query = "cough"
search_results, recs = recommend_medicines(query, df)

print(f"Search Results for: {query}")
display(search_results)

print("\nRecommended Alternatives (from substitutes):")
for r in recs:
    print("-", r)
```
